# Step 1 - Data Cleaning and Automation

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np

# Optional: show all columns when printing dataframes
pd.set_option('display.max_columns', None)

# Optional: display more rows
pd.set_option('display.max_rows', 20)
```

```python
# Load main Olist datasets
customers = pd.read_csv("data/raw/olist_customers_dataset.csv")
orders = pd.read_csv("data/raw/olist_orders_dataset.csv")
order_items = pd.read_csv("data/raw/olist_order_items_dataset.csv")
products = pd.read_csv("data/raw/olist_products_dataset.csv")

# Quick preview of each table
print("Customers Table")
display(customers.head())
print("Orders Table")
display(orders.head())
print("Order Items Table")
display(order_items.head())
print("Products Table")
display(products.head())
```

```python
# Check if customer_id is unique
print("Is customer_id unique?", customers['customer_id'].is_unique)

# Check if customer_unique_id is unique
print("Is customer_unique_id unique?", customers['customer_unique_id'].is_unique)
```

```python
# Shape of each table
print("Customers shape:", customers.shape)
print("Orders shape:", orders.shape)
print("Order Items shape:", order_items.shape)
print("Products shape:", products.shape)

# Missing values
print("\nMissing values per table:")
print("Customers:\n", customers.isnull().sum())
print("Orders:\n", orders.isnull().sum())
print("Order Items:\n", order_items.isnull().sum())
print("Products:\n", products.isnull().sum())

# Data types
print("\nData types per table:")
print("Customers:\n", customers.dtypes)
print("Orders:\n", orders.dtypes)
print("Order Items:\n", order_items.dtypes)
print("Products:\n", products.dtypes)
```

```python
# Step: Convert order date columns to datetime
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'], errors='coerce')
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'], errors='coerce')
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'], errors='coerce')
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])

# Quick check
orders.dtypes
```

```python
# Count missing timestamps in orders
missing_dates = orders[['order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date']].isnull().sum()
print("Missing timestamps per column:\n", missing_dates)
```

```python
# flags for pending approvals or deliveries
orders['approved_missing'] = orders['order_approved_at'].isnull()
orders['delivered_carrier_missing'] = orders['order_delivered_carrier_date'].isnull()
orders['delivered_customer_missing'] = orders['order_delivered_customer_date'].isnull()

# Quick preview
orders[['order_id', 'approved_missing', 'delivered_carrier_missing', 'delivered_customer_missing']].head()
```

```python
print("Orders Table")
display(orders.head())
```

```python
# Step: Check for negative or zero values in order_items
print("Number of items with price <= 0:", (order_items['price'] <= 0).sum())
print("Number of items with freight_value <= 0:", (order_items['freight_value'] <= 0).sum())
```

```python
# Inspect rows with negative freight_value
negative_freight = order_items[order_items['freight_value'] < 0]
display(negative_freight)
```

```python
# Check rows with missing product_category_name
missing_category = products[products['product_category_name'].isnull()]
display(missing_category)
```

```python
products['product_category_name'] = products['product_category_name'].fillna('unknown')

# Check missing values in product_category_name
print("Missing product categories:", products['product_category_name'].isnull().sum())

# Preview first 5 rows with 'unknown' category
missing_category = products[products['product_category_name'] == 'unknown']
display(missing_category.head())
```

```python
# Check how many missing values per numeric column
print(products[['product_name_lenght', 'product_description_lenght', 'product_photos_qty',
                'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']].isnull().sum())
```

```python
# Step: Fill missing numeric values in products table with median
numeric_cols = ['product_name_lenght', 'product_description_lenght', 'product_photos_qty',
                'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']

for col in numeric_cols:
    products[col] = products[col].fillna(products[col].median())

# Verify that there are no missing values left
print("Missing values after filling numeric columns:")
print(products[numeric_cols].isnull().sum())
```

```python
customers.to_csv("data/cleaned/olist_customers_clean.csv", index=False)
orders.to_csv("data/cleaned/olist_orders_clean.csv", index=False)
order_items.to_csv("data/cleaned/olist_order_items_clean.csv", index=False)
products.to_csv("data/cleaned/olist_products_clean.csv", index=False)

print("Cleaned tables saved in 'data/cleaned/' successfully!")
```

```python
# Create summary table: total orders and revenue per product category
order_items_products = order_items.merge(products, on='product_id', how='left')
summary = order_items_products.groupby('product_category_name').agg(
    total_orders=('order_id', 'count'),
    total_revenue=('price', 'sum')
).reset_index()

# Export cleaned tables + summary to one Excel file
with pd.ExcelWriter("data/cleaned/olist_full_report.xlsx", engine="openpyxl") as writer:
    customers.to_excel(writer, sheet_name="Customers", index=False)
    orders.to_excel(writer, sheet_name="Orders", index=False)
    order_items.to_excel(writer, sheet_name="Order Items", index=False)
    products.to_excel(writer, sheet_name="Products", index=False)
    summary.to_excel(writer, sheet_name="Summary", index=False)

print("All cleaned tables and summary saved in 'olist_full_report.xlsx'!")
```

# Exploratory Data Analysis (EDA)

```python
# Step: Basic statistics and overview of orders
print("Orders table shape:", orders.shape)
print("\nOrder status counts:")
print(orders['order_status'].value_counts())

print("\nOrder timestamps range:")
print("Earliest purchase:", orders['order_purchase_timestamp'].min())
print("Latest purchase:", orders['order_purchase_timestamp'].max())
```

```python
import matplotlib.pyplot as plt

# Step: Ensure order_purchase_timestamp is datetime
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])

# Extract month-year for grouping
orders['order_month'] = orders['order_purchase_timestamp'].dt.to_period('M')

# -----------------------------
# Monthly number of orders
monthly_orders = orders.groupby('order_month')['order_id'].count().reset_index()
monthly_orders.rename(columns={'order_id': 'num_orders'}, inplace=True)

plt.figure(figsize=(12,5))
plt.plot(monthly_orders['order_month'].astype(str), monthly_orders['num_orders'], marker='o', label='Number of Orders')
plt.xticks(monthly_orders['order_month'].astype(str)[::2], rotation=45)  # show every 2 months
plt.title("Number of Orders per Month")
plt.xlabel("Month-Year")
plt.ylabel("Number of Orders")
plt.grid(True)
plt.legend()
plt.show()

# -----------------------------
# Monthly revenue
# Merge orders with order_items to get prices
orders_items_merged = orders.merge(order_items, on='order_id', how='left')

monthly_revenue = orders_items_merged.groupby('order_month')['price'].sum().reset_index()
monthly_revenue.rename(columns={'price': 'total_revenue'}, inplace=True)

plt.figure(figsize=(12,5))
plt.plot(monthly_revenue['order_month'].astype(str), monthly_revenue['total_revenue'], marker='o', color='green', label='Total Revenue')
plt.xticks(monthly_revenue['order_month'].astype(str)[::2], rotation=45)
plt.title("Monthly Revenue")
plt.xlabel("Month-Year")
plt.ylabel("Revenue (BRL)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Step: Ensure delivery dates are datetime
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])

# Calculate delivery duration in days (only for delivered orders)
orders['delivery_duration_days'] = (orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']).dt.days

# Filter out orders that haven't been delivered yet (NaT)
delivered_orders = orders[orders['delivery_duration_days'].notnull()]

# Average delivery duration per month
avg_delivery_per_month = delivered_orders.groupby('order_month')['delivery_duration_days'].mean().reset_index()

# Plot average delivery duration over time
plt.figure(figsize=(12,5))
plt.plot(avg_delivery_per_month['order_month'].astype(str), avg_delivery_per_month['delivery_duration_days'], marker='o', color='orange')
plt.xticks(avg_delivery_per_month['order_month'].astype(str)[::2], rotation=45)
plt.title("Average Delivery Duration per Month")
plt.xlabel("Month-Year")
plt.ylabel("Delivery Duration (Days)")
plt.grid(True)
plt.show()
```
