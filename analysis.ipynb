# Step 1 - Data Cleaning and Automation

```python
# Import libraries
import pandas as pd
import numpy as np

# Display options for convenience
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)
```

```python
# Load main Olist datasets
customers = pd.read_csv("data/raw/olist_customers_dataset.csv")
orders = pd.read_csv("data/raw/olist_orders_dataset.csv")
order_items = pd.read_csv("data/raw/olist_order_items_dataset.csv")
products = pd.read_csv("data/raw/olist_products_dataset.csv")

# Quick preview
print("Customers Table"); display(customers.head())
print("Orders Table"); display(orders.head())
print("Order Items Table"); display(order_items.head())
print("Products Table"); display(products.head())
```

```python
# Check uniqueness & missing values
# ---------------------------
print("Is customer_id unique?", customers['customer_id'].is_unique)
print("Is customer_unique_id unique?", customers['customer_unique_id'].is_unique)

# Shape of each table
print("Customers shape:", customers.shape)
print("Orders shape:", orders.shape)
print("Order Items shape:", order_items.shape)
print("Products shape:", products.shape)

# Missing values
print("\nMissing values per table:")
print("Customers:\n", customers.isnull().sum())
print("Orders:\n", orders.isnull().sum())
print("Order Items:\n", order_items.isnull().sum())
print("Products:\n", products.isnull().sum())

# Data types
print("\nData types per table:")
print(customers.dtypes)
print(orders.dtypes)
print(order_items.dtypes)
print(products.dtypes)
```

```python
# Convert order date columns to datetime
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'], errors='coerce')
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'], errors='coerce')
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'], errors='coerce')
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])

# Flags for missing timestamps
orders['approved_missing'] = orders['order_approved_at'].isnull()
orders['delivered_carrier_missing'] = orders['order_delivered_carrier_date'].isnull()
orders['delivered_customer_missing'] = orders['order_delivered_customer_date'].isnull()

print("\nMissing timestamps per column:\n", orders[['order_approved_at', 
                                                    'order_delivered_carrier_date', 
                                                    'order_delivered_customer_date']].isnull().sum())
display(orders[['order_id', 'approved_missing', 'delivered_carrier_missing', 'delivered_customer_missing']].head())
```

```python
# Handle invalid order_items (outliers)
print("Original order_items shape:", order_items.shape)
order_items = order_items[(order_items['price'] > 0) & (order_items['freight_value'] >= 0)]
print("Cleaned order_items shape:", order_items.shape)
```

```python
# Handle missing product categories
products['product_category_name'] = products['product_category_name'].fillna('unknown')
print("Missing product categories after filling:", products['product_category_name'].isnull().sum())
```

```python
# Functionize numeric cleaning
def clean_numeric(df, cols):
    for col in cols:
        df[col] = df[col].fillna(df[col].median())
    return df

numeric_cols = ['product_name_lenght', 'product_description_lenght', 'product_photos_qty',
                'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']

products = clean_numeric(products, numeric_cols)
print("Missing values after cleaning numeric columns:")
print(products[numeric_cols].isnull().sum())
```

```python
# Validate joins
missing_orders = order_items[~order_items['order_id'].isin(orders['order_id'])]
print(f"Number of order_items with missing orders: {len(missing_orders)}")

# Optional: drop these if any
order_items = order_items[order_items['order_id'].isin(orders['order_id'])]
```

```python
# Save cleaned tables
customers.to_csv("data/cleaned/olist_customers_clean.csv", index=False)
orders.to_csv("data/cleaned/olist_orders_clean.csv", index=False)
order_items.to_csv("data/cleaned/olist_order_items_clean.csv", index=False)
products.to_csv("data/cleaned/olist_products_clean.csv", index=False)
print("Cleaned tables saved in 'data/cleaned/' successfully!")
```

```python
# Summary table for reports
order_items_products = order_items.merge(products, on='product_id', how='left')
summary = order_items_products.groupby('product_category_name').agg(
    total_orders=('order_id', 'count'),
    total_revenue=('price', 'sum')
).reset_index()

# Save all cleaned tables + summary to Excel
with pd.ExcelWriter("data/cleaned/olist_full_report.xlsx", engine="openpyxl") as writer:
    customers.to_excel(writer, sheet_name="Customers", index=False)
    orders.to_excel(writer, sheet_name="Orders", index=False)
    order_items.to_excel(writer, sheet_name="Order Items", index=False)
    products.to_excel(writer, sheet_name="Products", index=False)
    summary.to_excel(writer, sheet_name="Summary", index=False)

print("All cleaned tables and summary saved in 'olist_full_report.xlsx'!")
```

```python
# Enhanced dashboard summary (optional)
dashboard_summary = order_items_products.groupby('product_category_name').agg(
    total_orders=('order_id', 'count'),
    total_revenue=('price', 'sum'),
    avg_price=('price', 'mean'),
    median_price=('price', 'median')
).reset_index()

dashboard_summary.to_excel("data/cleaned/olist_dashboard_summary.xlsx", index=False)
print("Dashboard summary saved successfully!")
```

```python
import matplotlib.pyplot as plt

# Step: Ensure order_purchase_timestamp is datetime
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])

# Extract month-year for grouping
orders['order_month'] = orders['order_purchase_timestamp'].dt.to_period('M')

# -----------------------------
# Monthly number of orders
monthly_orders = orders.groupby('order_month')['order_id'].count().reset_index()
monthly_orders.rename(columns={'order_id': 'num_orders'}, inplace=True)

plt.figure(figsize=(12,5))
plt.plot(monthly_orders['order_month'].astype(str), monthly_orders['num_orders'], marker='o', label='Number of Orders')
plt.xticks(monthly_orders['order_month'].astype(str)[::2], rotation=45)  # show every 2 months
plt.title("Number of Orders per Month")
plt.xlabel("Month-Year")
plt.ylabel("Number of Orders")
plt.grid(True)
plt.legend()
plt.show()

# -----------------------------
# Monthly revenue
# Merge orders with order_items to get prices
orders_items_merged = orders.merge(order_items, on='order_id', how='left')

monthly_revenue = orders_items_merged.groupby('order_month')['price'].sum().reset_index()
monthly_revenue.rename(columns={'price': 'total_revenue'}, inplace=True)

plt.figure(figsize=(12,5))
plt.plot(monthly_revenue['order_month'].astype(str), monthly_revenue['total_revenue'], marker='o', color='green', label='Total Revenue')
plt.xticks(monthly_revenue['order_month'].astype(str)[::2], rotation=45)
plt.title("Monthly Revenue")
plt.xlabel("Month-Year")
plt.ylabel("Revenue (BRL)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Step: Ensure delivery dates are datetime
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])

# Calculate delivery duration in days (only for delivered orders)
orders['delivery_duration_days'] = (orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']).dt.days

# Filter out orders that haven't been delivered yet (NaT)
delivered_orders = orders[orders['delivery_duration_days'].notnull()]

# Average delivery duration per month
avg_delivery_per_month = delivered_orders.groupby('order_month')['delivery_duration_days'].mean().reset_index()

# Plot average delivery duration over time
plt.figure(figsize=(12,5))
plt.plot(avg_delivery_per_month['order_month'].astype(str), avg_delivery_per_month['delivery_duration_days'], marker='o', color='orange')
plt.xticks(avg_delivery_per_month['order_month'].astype(str)[::2], rotation=45)
plt.title("Average Delivery Duration per Month")
plt.xlabel("Month-Year")
plt.ylabel("Delivery Duration (Days)")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# --- Ensure datetime columns ---
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])

# --- Merge orders with customer state ---
orders_state = orders.merge(
    customers[['customer_id', 'customer_state']],
    on='customer_id',
    how='left'
)

# --- Orders per state ---
orders_per_state = (
    orders_state
    .groupby('customer_state')['order_id']
    .count()
    .reset_index()
    .rename(columns={'order_id': 'num_orders'})
    .sort_values(by='num_orders', ascending=False)
)

# --- Delivered orders only ---
delivered_orders_state = orders_state[
    orders_state['order_status'] == 'delivered'
].copy()

# --- Delivery duration ---
delivered_orders_state['delivery_duration_days'] = (
    delivered_orders_state['order_delivered_customer_date']
    - delivered_orders_state['order_purchase_timestamp']
).dt.days

# --- Average delivery duration per state ---
avg_delivery_per_state = (
    delivered_orders_state
    .groupby('customer_state')['delivery_duration_days']
    .mean()
    .reset_index()
)

# Keep same state order
state_order = orders_per_state['customer_state']
avg_delivery_per_state = (
    avg_delivery_per_state
    .set_index('customer_state')
    .reindex(state_order)
    .reset_index()
)

# --- Side-by-side plots ---
fig, axes = plt.subplots(1, 2, figsize=(18, 5))

# Orders count
axes[0].bar(orders_per_state['customer_state'], orders_per_state['num_orders'])
axes[0].set_title("Number of Orders per State")
axes[0].set_xlabel("State")
axes[0].set_ylabel("Number of Orders")
axes[0].tick_params(axis='x', rotation=45)

# Delivery duration
axes[1].bar(avg_delivery_per_state['customer_state'], avg_delivery_per_state['delivery_duration_days'])
axes[1].set_title("Average Delivery Duration per State")
axes[1].set_xlabel("State")
axes[1].set_ylabel("Delivery Duration (Days)")
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

### Cohort Retention Analysis

```python
# Merge orders with customer_unique_id
orders_customers = orders.merge(
    customers[['customer_id', 'customer_unique_id']],
    on='customer_id',
    how='left'
)

# Number of orders per unique customer
orders_per_customer = (
    orders_customers
    .groupby('customer_unique_id')['order_id']
    .nunique()
    .reset_index()
    .rename(columns={'order_id': 'num_orders'})
)

# Label customers
orders_per_customer['customer_type'] = np.where(
    orders_per_customer['num_orders'] == 1,
    'One-time',
    'Repeat'
)

orders_per_customer['customer_type'].value_counts()
# Revenue per order
order_revenue = (
    order_items
    .groupby('order_id')['price']
    .sum()
    .reset_index()
)

# Attach revenue to customers
customer_revenue = (
    orders_customers
    .merge(order_revenue, on='order_id', how='left')
    .merge(
        orders_per_customer[['customer_unique_id', 'customer_type']],
        on='customer_unique_id',
        how='left'
    )
)

# Revenue by customer type
revenue_by_type = (
    customer_revenue
    .groupby('customer_type')['price']
    .sum()
    .reset_index()
    .rename(columns={'price': 'total_revenue'})
)

revenue_by_type
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.bar(revenue_by_type['customer_type'], revenue_by_type['total_revenue'])
plt.title("Revenue Contribution by Customer Type")
plt.xlabel("Customer Type")
plt.ylabel("Total Revenue")
plt.show()
```

```python
orders_cohort = orders.merge(
    customers[['customer_id', 'customer_unique_id']],
    on='customer_id',
    how='left'
)

orders_cohort.head()
```

```python
# Extract purchase month (year-month)
orders_cohort['order_month'] = (
    orders_cohort['order_purchase_timestamp']
    .dt.to_period('M')
)

# First purchase month per customer
first_purchase = (
    orders_cohort
    .groupby('customer_unique_id')['order_month']
    .min()
    .reset_index()
    .rename(columns={'order_month': 'cohort_month'})
)

first_purchase.head()
```

```python
# Merge cohort month back into orders
orders_cohort = orders_cohort.merge(
    first_purchase,
    on='customer_unique_id',
    how='left'
)

# Calculate number of months since first purchase
orders_cohort['cohort_index'] = (
    orders_cohort['order_month'] - orders_cohort['cohort_month']
).apply(lambda x: x.n)

orders_cohort[['customer_unique_id', 'order_month', 'cohort_month', 'cohort_index']].head()
```

```python
# Count unique customers per cohort and cohort_index
cohort_counts = (
    orders_cohort
    .groupby(['cohort_month', 'cohort_index'])['customer_unique_id']
    .nunique()
    .reset_index()
)

cohort_counts.head()
```

```python
# Get cohort sizes (Month 0 counts)
cohort_sizes = (
    cohort_counts[cohort_counts['cohort_index'] == 0]
    [['cohort_month', 'customer_unique_id']]
    .rename(columns={'customer_unique_id': 'cohort_size'})
)

# Merge sizes back
cohort_retention = cohort_counts.merge(
    cohort_sizes,
    on='cohort_month',
    how='left'
)

# Calculate retention rate
cohort_retention['retention_rate'] = (
    cohort_retention['customer_unique_id'] / cohort_retention['cohort_size']
)

cohort_retention.head()
```

# Step 3 - Statistical Modeling

```python
# Load payments table
payments = pd.read_csv("data/raw/olist_order_payments_dataset.csv")

payments.head()
```

```python
# Sort orders by customer and time
orders_model = orders_cohort.sort_values(
    ['customer_unique_id', 'order_purchase_timestamp']
)

# Label repeat purchase (1 if customer has a future order)
orders_model['is_repeat'] = (
    orders_model
    .groupby('customer_unique_id')['order_purchase_timestamp']
    .shift(-1)
    .notna()
    .astype(int)
)

orders_model[['customer_unique_id', 'order_purchase_timestamp', 'is_repeat']].head(10)
```

```python
orders_model['is_repeat'].value_counts(normalize=True) * 100
```

```python
orders_model['is_repeat'].value_counts()
```

```python
# Aggregate customer-level features (corrected)
customer_features = (
    orders
    .merge(customers[['customer_id', 'customer_unique_id']], on='customer_id', how='left')
    .merge(order_items, on='order_id', how='left')
    .merge(payments, on='order_id', how='left')
    .groupby('customer_unique_id')
    .agg(
        num_orders=('order_id', 'nunique'),
        total_spent=('payment_value', 'sum'),
        avg_order_value=('payment_value', 'mean'),
        first_purchase=('order_purchase_timestamp', 'min'),
        last_purchase=('order_purchase_timestamp', 'max')
    )
    .reset_index()
)

# Target variable: repeat customer
customer_features['is_repeat'] = (customer_features['num_orders'] > 1).astype(int)

customer_features.head()
```

```python
# Time based features
customer_features['customer_lifetime_days'] = (
    customer_features['last_purchase'] - customer_features['first_purchase']
).dt.days

customer_features['customer_lifetime_days'] = customer_features['customer_lifetime_days'].fillna(0)

customer_features.head()
```

```python
features = [
    'total_spent',
    'avg_order_value',
    'customer_lifetime_days'
]

X = customer_features[features]
y = customer_features['is_repeat']
```

```python
# Fill missing values with business logic
X = X.copy()

X['total_spent'] = X['total_spent'].fillna(0)

X['avg_order_value'] = X['avg_order_value'].fillna(X['total_spent'])

X['customer_lifetime_days'] = X['customer_lifetime_days'].fillna(0)

# Double-check
X.isna().sum()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(
    class_weight='balanced',
    max_iter=1000
)

log_model.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import classification_report, roc_auc_score

y_pred = log_model.predict(X_test_scaled)
y_prob = log_model.predict_proba(X_test_scaled)[:, 1]

print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_prob))
```

```python
coef_df = pd.DataFrame({
    'feature': features,
    'coefficient': log_model.coef_[0],
    'odds_ratio': np.exp(log_model.coef_[0])
}).sort_values(by='odds_ratio', ascending=False)

coef_df
```
